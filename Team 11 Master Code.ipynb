```python
#Importing Data

file='WDIW Dataset (1).xlsx'
africa=pd.read_excel(file)

africa
```

<strong>Data cleaning for analysis

```python
#filtering columns

africa_1=africa.loc[africa['Cool Name']=='Sideswipe',[
'Country Name',
'Surface area (sq. km)',
'GDP (current US$)',
'GDP growth (annual %)',
'Tax revenue (% of GDP)',
'Agriculture, forestry, and fishing, value added (% of GDP)',
'Industry (including construction), value added (% of GDP)',
'Services, value added (% of GDP)',
'Merchandise trade (% of GDP)',
'Access to electricity (% of population)',
'Access to electricity, rural (% of rural population)',
'Access to electricity, urban (% of urban population)',
'Government expenditure on education, total (% of government expenditure)',
'Military expenditure (% of GDP)',
'Population density (people per sq. km of land area)',
'Population ages 0-14 (% of total population)',
'Population ages 15-64 (% of total population)',
'Population ages 65 and above (% of total population)',
'Population, female',
'Population, male',
'Population, total',
'Population growth (annual %)']]

africa_1
```

```python
#Re_indexing
africa_1.reset_index(inplace = True) 
del africa_1['index']
```

```python
africa_1
```

<strong>New variable and missing values

```python
# Adding new columns for new rates
#In order to evaluate the contribution to GDP per person/square_meter

africa_1['GDP_per_Sqr_meter'] = africa_1['GDP (current US$)']/africa_1['Surface area (sq. km)']
africa_1['People_with_elecp.sq.m'] = africa_1['Population density (people per sq. km of land area)']/africa_1['Access to electricity (% of population)']
africa_1['GDP_per_Person'] = africa_1['GDP (current US$)']/africa_1['Population, total']


africa_1
```

```python
#mean of historical data for Government expenditure on education
#historical data found on world bank 
benin_2016=18.791
benin_2015=17.48
benin_2014=22.228
burk_2016=17.853
burk_2015=18.027
burk_2014=19.368
mean_benin=(benin_2014+benin_2015+benin_2016)/3
mean_burk=(burk_2014+burk_2015+burk_2016)/3
print(mean_benin)
print(mean_burk)
```

```python
# imputing missing Value on Government expenditure on education

africa_1.loc[1,'Government expenditure on education, total (% of government expenditure)'] = 19.499999
africa_1.loc[2,'Government expenditure on education, total (% of government expenditure)'] =18.416

#Because of the lack of historical data on education spending for 
#some of the countries in the region we were analyzing,
#we realized that most of these countries' education spending 
#almost is none existant or never reported as frequently as possible.
#In terms of priority, some of these governments have some more 
#significant problems to solve; hence, they overlook education.
```

```python
# imputing missing Value Total tax Revenue %
#missing data found from cia.gov

africa_1.loc[1,'Tax revenue (% of GDP)'] = 17.1
africa_1.loc[5,'Tax revenue (% of GDP)']=14.6
africa_1.loc[6,'Tax revenue (% of GDP)']=25.3
africa_1.loc[7,'Tax revenue (% of GDP)']=11.2
africa_1.loc[10,'Tax revenue (% of GDP)']=17.6
africa_1.loc[11,'Tax revenue (% of GDP)']=20.3
africa_1.loc[12,'Tax revenue (% of GDP)']=20.3
africa_1.loc[13,'Tax revenue (% of GDP)']=16.6
africa_1.loc[15,'Tax revenue (% of GDP)']=16.9
africa_1.loc[17,'Tax revenue (% of GDP)']=3.4
africa_1.loc[18,'Tax revenue (% of GDP)']=26.2
africa_1.loc[19,'Tax revenue (% of GDP)']=19.6
africa_1.loc[20,'Tax revenue (% of GDP)']=39.6
africa_1.loc[21,'Tax revenue (% of GDP)']=15.6
```

```python
#after checking the data with no outlier and with few available data available to us
#we decided to take the column mean and impute our 2 missing values for this section

rural_elec = africa_1['Access to electricity, rural (% of rural population)'].mean()
africa_1['Access to electricity, rural (% of rural population)'] = africa_1['Access to electricity, rural (% of rural population)'].fillna(rural_elec)

print(africa_1['Access to electricity, rural (% of rural population)'].isnull().any())
```

```python
#comparing population of male to female in the region 
#per country 
y = africa_1['Country Name']
x1 = africa_1['Population, male']
x2 = africa_1['Population, female']

fig, axes = plt.subplots(ncols=2, sharey=True)
axes[0].barh(y, x1, align='center', color='black',height=1)
axes[0].set(title='Male')
axes[1].barh(y, x2, align='center', color='g',height=1)
axes[1].set(title='Female')
axes[0].invert_xaxis()
axes[0].yaxis.tick_right()
axes[0].set(yticks=y, yticklabels=[])
for yloc, state in zip(y, y):
    axes[0].annotate(state, (0.5, yloc), xycoords=('figure fraction', 'data'),
                     ha='center', va='center')


fig.tight_layout()
fig.subplots_adjust(wspace=0.8)
plt.savefig('male_female_dist')
```

```python
#to determine how well the country is doing when compared to its size
sns.lmplot(x='GDP_per_Sqr_meter',y='GDP (current US$)',data=africa_1,height=8)

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.01, point['y'], str(point['val']))

label_point(africa_1['GDP_per_Sqr_meter'], africa_1['GDP (current US$)'], africa_1['Country Name'], plt.gca())
plt.savefig('gdp_per_sq_met')
```

```python
#comparing Gdp per capita of the region 
#Gdp per capita in Nigeria is low , but total Gdp is high 
#while in the Seychelles is the opposite

# Analysis of GDP performance per Country

#Setting up Variables (three dimensions and a variable for random colors)
x=africa_1['GDP_per_Person']
y= africa_1['GDP (current US$)']
z=africa_1['Surface area (sq. km)']
w=[np.random.rand(3,) for iter in range(22)]

#Setting up Size and Font of plot
fig, ax = plt.subplots(figsize=(9,6))
plt.rcParams.update({'font.size':9})

#Creating a Bubble chart
plt.scatter(x,y,s=z*0.004,alpha=0.5,c=w) #s is for global size

plt.title('GDP Performance')
plt.xlabel('GDP per Capita')
plt.ylabel('Total GDP')

#Creating a legend for the chart
country_list=list(africa_1['Country Name'])
color_list=list(w)
l=[]
for i in range(0,len(africa_1.index)):
    l.append(mpatches.Patch(color=color_list[i],alpha=0.5,label=country_list[i]))
    
plt.legend(handles=l,loc=(1.05,0))   

plt.savefig('GDP Comparison',dpi = 100, bbox_inches='tight')

plt.show()
```
